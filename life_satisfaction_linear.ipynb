```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Download and prepare the data
lifesat = pd.read_csv("./lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
print("inputs X:", type(X), X.shape)
print("targets y:", type(y), y.shape)

# Visualize the data
lifesat.plot(kind="scatter", grid=True, x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])


# Select a linear model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Print the equation of the model
m = model.coef_[0]  # slope
b = model.intercept_  # intercept
print("slope:", m)
print("intercept:", b)

# Make a prediction for Cyprus
X_new = [[37_655.2]]  # Cyprus' GDP per capita in 2020
print("predicted value:", model.predict(X_new))  # outputs [[6.30165767]]

# Plot the regression line
plt.plot(X, model.predict(X), "r-")  # red line
plt.plot(X_new, model.predict(X_new), "ro")  # red dot for Cyprus
plt.text(25000, 5.0, r"Prediction = 6.30", fontsize=14, color="r")  # label for Cyprus
plt.show()
```
